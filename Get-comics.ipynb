```python
# Getting links to each image of comics
url = "https://read-comic.com/old-man-logan-01/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
links = []
  
ilist = soup.findAll('img')
for im in ilist:
    imsrc = im['src']
    links.append(imsrc)
```
